```python
import pandas as pd
import requests
```

## 1. Obtener los datos desde las API's.

### Lista de pasajeros en el 2016 convertida a df:

```python
#Acceso a la API usando la llave correcta:
r1 = requests.get('http://analytics.deacero.com/Api/GetApi/ApiPasajeros2016/ecfb5fc7-0932-590f-832c-6d6055f2be07')
#Obtención de la información en JSON:
lista_pasajeros_2016 = (r1.json())
#Creación del DataFrame de Pandas:
df_pasajeros_2016 = pd.DataFrame(lista_pasajeros_2016)
df_pasajeros_2016
```

### Lista de pasajeros en el 2017 convertida a df:

```python
r2 = requests.get('http://analytics.deacero.com/Api/GetApi/ApiPasajeros2017/faabd632-cc39-552d-a68b-02de4242f636')
lista_pasajeros_2017 = (r2.json())
df_pasajeros_2017 = pd.DataFrame(lista_pasajeros_2017)
df_pasajeros_2017
```

### Lista de vuelos en el 2016 convertida a df:

```python
r3 = requests.get('http://analytics.deacero.com/Api/GetApi/ApiVuelos2016/9ea3b836-6938-52dc-9626-a8e35db81dd5')
lista_vuelos_2016 = (r3.json())
df_vuelos_2016 = pd.DataFrame(lista_vuelos_2016)
df_vuelos_2016
```

### Lista de vuelos en el 2017 convertida a df:

```python
r4 = requests.get('http://analytics.deacero.com/Api/GetApi/ApiVuelos2017/fc126260-1cf8-5a46-995d-ba639ff5868b')
lista_vuelos_2017 = (r4.json())
df_vuelos_2017 = pd.DataFrame(lista_vuelos_2017)
df_vuelos_2017
```

### Lista de aerolíneas convertida a df:

```python
r5 = requests.get('http://analytics.deacero.com/Api/GetApi/ApiLineaAerea/1a8d9e13-ce30-50fc-bf34-6490eb799a75')
lista_aerolineas = (r5.json())
df_aerolineas = pd.DataFrame(lista_aerolineas)
df_aerolineas
```

## 2. Se tiene un requerimiento de análisis de las fuentes de datos de pasajeros y viajes. Como puede observarse, las listas se han estado llevando por año. En esta primera parte del reto se pide unir cada conjunto de datos en una sola lista.

### Concatenación de los df's de pasajeros del 2016 y 2017:

```python
#Se utiliza la funcion concat() de Pandas para unir la información de los dos años en un solo DataFrame, ya que concat() nos
#permite unir tablas con información similar a lo largo de las filas o de las columnas. En este caso se une sobre las columnas.
df_pasajeros_total = pd.concat([df_pasajeros_2016, df_pasajeros_2017], ignore_index=True) 
df_pasajeros_total
```

### Concatenación de los df's de vuelos del 2016 y 2017:

```python
df_vuelos_total = pd.concat([df_vuelos_2016, df_vuelos_2017], ignore_index=True)
df_vuelos_total
```

## 3. Se requiere relacionar las listas de vuelos y pasajeros, que permitan analizar el perfil del pasajero por cada vuelo efectuado.

```python
#Usamos la función join() de Pandas para unir las dos tablas basándonos en las columnas que tienen información en común. Dado
#que desconozco si todos los pasajeros están incluidos en los vuelos de la otra lista, y desconozco por igual si los vuelos
#fueron tomados únicamente por los pasajeros que se me dan en la lista y no por algunos otros no listados, se procede
#a hacer un left join y después un right join para descubrir esta información.
#Tras hacer ambos joins, me doy cuenta de que en efecto todos los vuelos fueron tomados por los pasajeros que tengo en mi tabla,
#pero no todos los vuelos de los pasajeros están registrados en la tabla de vuelos, y dado que mis pasajeros incluyen a todos
#los vuelos tomo esta tabla como la primera y hago un left join para incluir a los pasajeros que no tomaron los vuelos listados
#en la otra tabla.

# Nótese que también es posible usar: pd.merge(df_pasajeros_total, df_vuelos_total, left_on='ID_Pasajero', right_on='Cve_Cliente', how='left').drop_duplicates(ignore_index=True) 
#Pero esto causará que se dupliquen las columnas en donde se realiza el join, es por eso que se utiliza join().
df_vuelos_pasajeros = df_pasajeros_total.join(df_vuelos_total.set_index(['Cve_Cliente']), on=['ID_Pasajero']).drop_duplicates(ignore_index=True)
df_vuelos_pasajeros
```

## 4. Ahora se requiere que los datos consolidados de los vuelos y pasajeros se puedan unir con los datos de las Líneas Aéreas. En el caso de que la línea aérea no se pueda relacionar con la de vuelos y pasajeros se deberá indicar que se trata de “Otra” y finalmente se deberá dejar únicamente las columnas de fecha del viaje, clase, precio, ruta, edad y línea aérea.

```python
# Al igual que en el caso anterior, se utiliza la función join() para unir ambas tablas, ya que de esta manera se evita obtener
#una duplicación de las columnas donde se realiza el join.
#Si se presta atención, en este caso se utiliza el parámetro "how" para especificar el tipo de join, en el caso anterior este
#parametro no es necesario, ya que por default la funcion join() realiza left joins.
#Decidí hacer esto para mantener el nombre "Code" en mi tabla, pues join() utiliza el nombre de la tabla de la izquierda
#y se deshace del de la tabla de la derecho. Esto es porque aunque nuestras tablas comparten la misma información en dicha
#columna, el nombre de las columnas es diferente.
df_vuelos_pasajeros_aerolineas = df_aerolineas.join(df_vuelos_pasajeros.set_index(['Cve_LA']), how = 'right', on=['Code']).drop_duplicates(ignore_index=True)
df_vuelos_pasajeros_aerolineas
```

```python
#Para sustituir los valores nulos en la columna "Linea_Aerea", se usa el seleccionador con el nombre de la columna en la cual se
#se desean sustituir los valores nulos y sobre ésta se utiliza la función fillna() con el parametro "inplace = True".
#Este parámetro se agrega para que los cambios surtan efecto sobre nuestra tabla original y no tengamos sólo una vista temporal.
df_vuelos_pasajeros_aerolineas["Linea_Aerea"].fillna("Otra", inplace = True)
#Para dejar únicamente las columnas que se nos piden utilizamos el seleccionador, que consta de una lista de las columnas que
#queremos, y las asignamos a nuestro DataFrame original
df_vuelos_pasajeros_aerolineas = df_vuelos_pasajeros_aerolineas[['Viaje', 'Clase', 'Precio', 'Ruta', 'Edad', "Linea_Aerea"]]
df_vuelos_pasajeros_aerolineas
```

## 5. Por último, se requiere el promedio semestral (el primer semestre es de Ene - Jun y el segundo es de Jul - Dic) del precio agrupado por Año, Clase, Ruta y Línea Aérea como columnas.

```python
#Para poder llevar a cabo esta tarea, es necesario que nuestra columna "Viaje", la cual guarda la fecha del viaje, sea
#convertida a date time format, ya que por default el tipo de este dato es un objeto. La función to_datetime() de pandas nos 
#ayudará con esto:
df_vuelos_pasajeros_aerolineas['Viaje']=pd.to_datetime(df_vuelos_pasajeros_aerolineas['Viaje'])
#Esto nos ayudará también a ordenar nuestra tabla por fecha, y procedemos a hacerlo. Se utiliza reset_index() con drop = True
#para que los cambios se hagan sobre el índice original
df_vuelos_pasajeros_aerolineas = df_vuelos_pasajeros_aerolineas.sort_values(by=['Viaje']).reset_index(drop=True)
#Dado que no tiene caso mantener las columnas con ningún dato aparte de la edad del pasajero (estos son los pasajeros que no se
#encontraban en la lista de vuelos), se procede a eliminar estas entradas usando la función dropna()
df_vuelos_pasajeros_aerolineas.dropna(inplace=True)
df_vuelos_pasajeros_aerolineas
```

### Promedio semestral agrupado por clase:

```python
#Se utiliza la función grouby() de pandas para agrupar por año, semestre y clase del vuelo. En el caso de las dechas dentro
#de groupby(), además se utiliza Grouper(), que nos ayudará a trabajar con las fechas. Después de agrupar se selecciona el
#el precio y se utiliza la función mean() redondeada a dos decimas, lo que nos dará la media, o promedio.
promedio_precio_clase  = df_vuelos_pasajeros_aerolineas.groupby([pd.Grouper(key='Viaje', freq='Y'), pd.Grouper(key='Viaje', freq='6MS'), "Clase"])["Precio"].mean().round(2)
promedio_precio_clase
```

### Promedio semestral agrupado por ruta:

```python
#Se utiliza la función grouby() de pandas para agrupar por año, semestre y ruta. En el caso de las dechas dentro
#de groupby(), además se utiliza Grouper(), que nos ayudará a trabajar con las fechas. Después de agrupar se selecciona el
#el precio y se utiliza la función mean() redondeada a dos decimas, lo que nos dará la media, o promedio.
promedio_precio_ruta  = df_vuelos_pasajeros_aerolineas.groupby([pd.Grouper(key='Viaje', freq='Y'), pd.Grouper(key='Viaje', freq='6MS'), "Ruta"])["Precio"].mean().round(2)
promedio_precio_ruta
```

### Promedio semestral agrupado por línea aérea:

```python
#Se utiliza la función grouby() de pandas para agrupar por año, semestre y aerolínea. En el caso de las dechas dentro
#de groupby(), además se utiliza Grouper(), que nos ayudará a trabajar con las fechas. Después de agrupar se selecciona el
#el precio y se utiliza la función mean() redondeada a dos decimas, lo que nos dará la media, o promedio.
promedio_precio_aerolinea  = df_vuelos_pasajeros_aerolineas.groupby([pd.Grouper(key='Viaje', freq='Y'), pd.Grouper(key='Viaje', freq='6MS'), "Linea_Aerea"])["Precio"].mean().round(2)
promedio_precio_aerolinea
```
